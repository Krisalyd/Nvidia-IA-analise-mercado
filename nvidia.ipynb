# Qual a relação entre Nvidia, Inteligência Artificial e a grande disparada no valor de mercado?

Desde 1993, a Nvidia teve como principal atividade o desenvolvimento de chips programáveis, também conhecido como microcontroladores. A entrada da Nvidia no mercado gerou uma forte luta contra as concorrentes AMD e Intel.

A empresa era popular por seu envolvimento com o universo gamer com a criação de GPUs de alta performance e sua capacidade de formar grandes parcerias para fornecer este produto, como Microsoft, Google e Sony.

## Mas como a Nvidia se consolidou no campo de Inteligência Artificial?

Em meados de 2019, muitos pesquisadores da área de IA descobriram que as GPUs da NVIDIA poderiam ser a base adequada para o desenvolvimento de projetos de IA devida sua alta capacidade de paralelismo, que é a possibilidade de executar múltiplos processos simultaneamente. Nesta mesma época, as receitas relacionadas a IA da Nvidia chegaram ao patamar de US$ 2,9 bilhões.

```python
import yfinance as yf
import plotly.express as px
import pandas as pd

nvda = yf.Ticker("NVDA")
amd = yf.Ticker("AMD")
intc = yf.Ticker("INTC")
```

Logo abaixo temos o gráfico com o preço da ação da Nvidia e suas concorrentes diretas, AMD e Intel desde sua listagem na bolsa de valores e de fato é possível notar as expressivas disparadas de preço da Nvidia a partir de 2019, devido aos aumento das receitas reportadas e também o aumenta da confiança dos investidores neste novo tipo de empreendimento.

```python
nvda_historico = nvda.history(period="max")
amd_historico = amd.history(period="max")
intc_historico = intc.history(period="max")

nvda_historico.insert(0, "Ticker", "Nvidia", True)
amd_historico.insert(0, "Ticker", "AMD", True)
intc_historico.insert(0, "Ticker", "Intel", True)

concat_historico = pd.concat([nvda_historico, amd_historico, intc_historico])

preco_historico = px.line(x=concat_historico.index, y = concat_historico["Close"], labels = {"x":"Período", "y": "Preço da Ação (US$)"}, color = concat_historico["Ticker"])
preco_historico.show("png")

```

Suas especificações únicas e dominância no mercado permitiram que a marca histórica da empresa de valor de mercado de US$ 1,97 Trilhão fosse alcançada.
